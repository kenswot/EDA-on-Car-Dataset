## Working with a Car Datasets
Here. The data of different cars is given with specification


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
car = pd.read_csv('Cars Data1.csv')
```

```python
car.head()
```

```python
car.shape
```

## Cleaning the Data
Find the Null value in the datasets. And if there are any fill it with the mean column,

```python
car.isnull().sum()
```

```python
car['Cylinders'].fillna(car['Cylinders'].mean(), inplace = True)
```

```python
car
```

```python
car['Make'].value_counts()
```

```python
car['MPG_City'] = car['MPG_City'].apply(lambda x:x+3)
```
